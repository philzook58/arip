# Is logic one thing?
A confusion that occurs is thinking that there is one thing that is logic.

You may have encountered boolean logic before and the concept of true/false. Digital circuits, and the ability to equational manipulate boolean expressions in a manner similar to algebraic ones.

There is a conflation of a lot of distinct ideas in these notions.

Some of the problem occurs in extrapolating things we that everyone pretty much agrees about finitary objects (and really, modest finitary at that) to infinitary (or ludicrously large) ones.




# Judgements

Judgements are pieces of knowledge.

An easy confusion to fall into is to conflate the judgement with the thing the judgement is about. We can discuss whether a ballon is red without knowing if the the balloon is actually red. In this way, the judgement that a formula is true is not the same as the formula itself.

We can easily distinguish the two in a programming context because they just plain aren't the same data structure.


Some common judgements may be "t has type A in context Gamma" which is written in punctuation as `Gamma |- t : A`.

Other judgements might be `A is true` or `A is not true`

# Models
Proof is about playing the rules of a syntactic game. The only reason we are interesting in these rules is we have an intuitive feeling they are saying something about interesting objects.

There is a certain circularity of the concept of model. We claim that our syntactic games make sense because we can build models, which are set theoretic things. But the sticky and controversial parts of set theory about the infinite can only be carefully dealt with by considering them syntactic games also.
